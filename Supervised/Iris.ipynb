$\textbf{Steps To Be followed When Applying an Algorithm}$ $ \\ $
+ Split the dataset into training and testing dataset. The testing dataset is generally smaller than training one as it will help in training the model better. $ \\ $
+ Select any algorithm based on the problem. $ \\ $
+ Then pass the training dataset to the algorithm to train it. We use the .fit() method $ \\ $
+ Then pass the testing data to the trained algorithm to predict the outcome. We use the .predict() method. $ \\ $
+ We then check the accuracy by passing the predicted outcome and the actual output to the model.$ \\ $

$\textbf{Splitting The Data into Training And Testing Dataset}$

```python
train, test = train_test_split(data, test_size = 0.2)# in this our main data is split into train and test
# the attribute test_size=0.3 splits the data into 80% and 20% ratio. train=80% and test=20%
print("Train shape: ", train.shape)
print("Test shape: ", test.shape)
print("\n \n")

train_x = train.loc[:, data.columns != "Species"]
print("Train_x: \n", train_x.head(2))
train_y = train.loc[:, data.columns == "Species"]
print("Train_y: \n", train_y.head(2))

print("\n \n")
test_x = test.loc[:, data.columns != "Species"]
print("Test_x: \n", test_x.head(2))
test_y = test.loc[:, data.columns == "Species"]
print("Test_y: \n", test_y.head(2))
```

$\textbf{K-Neighbors Classifier - sklearn}$

```python
train_score = []
test_score = []
range_k = range(2,11,1)

for _ in range_k:
    model = KNeighborsClassifier(n_neighbors = _, 
                                 weights= 'uniform',
                                 algorithm= 'brute') #this examines _ neighbours for putting the new data into a class
    model.fit(train_x, np.ravel(train_y))
    prediction = model.predict(test_x)

    train_score.append(model.score(train_x, train_y))
    test_score.append(model.score(test_x, test_y))

print("TRAIN SCORE: \n", train_score)
print("TEST SCORE: \n", test_score)

fig = sns.scatterplot(y = train_score, x = range_k, label = "Train Score", color = "r")
sns.scatterplot(y = test_score, x = range_k, label = "Test Score", color = "g", ax= fig)
plt.show()
```
